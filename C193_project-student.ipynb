```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df=pd.read_csv("cardiovascular.csv")
df=df.dropna()
age = df.groupby('age')['cholesterol'].mean().reset_index()
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it
label = age['age']
value = age['cholesterol']

fig = plt.subplots(figsize=(19,8))

plt.plot(label, value, label = "Average cholesterol as per date" , linewidth=3.0)
plt.xlabel('age')
plt.xticks(rotation='vertical')

plt.ylabel('cholesterol')

plt.title('Average cholesterol as per date', fontsize=20)

plt.legend()

plt.show()
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
systolic = df.groupby('age')['systolic_blood_pressure'].max().reset_index()
systolic
diastolic = df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
diastolic
sl = systolic['age']
sval = systolic['systolic_blood_pressure']

fig1 = plt.subplots(figsize=(19,8))

plt.plot(sl, sval, label = "Average systolic date" , linewidth=3.0)

dl = diastolic['age']
dval = diastolic['diastolic_blood_pressure']

fig1 = plt.subplots(figsize=(19,8))

plt.plot(dl, dval, label = "Average diastolic date" , linewidth=3.0)
plt.xlabel('age')
plt.xticks(rotation='vertical')

plt.ylabel('blood_pressure')

plt.title('Average systolic and diastolic as per date', fontsize=20)

plt.legend()

plt.show()
```
